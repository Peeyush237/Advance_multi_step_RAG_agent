```python
from langgraph.graph import StateGraph, START, END
from langgraph.types import Command, interrupt
from typing import TypedDict
from langgraph.checkpoint.memory import MemorySaver

memory = MemorySaver()  ## to   maintain the memory
```

```python
class state(TypedDict):
    value :str

def node_a(state:state):
    print("Node A")
    return Command(
        goto="node_b",
        update={

            "value": state["value"]+"a"
        }
    )

def node_b(state:state):
    print("NODE B")

    human_response = interrupt("do you want to go to C or D? type C or D")
    print("human review values:", human_response)

    if(human_response == "C"):
        return Command(
            goto="node_c",
            update={

                "value": state["value"] + "b"
            }
        )
    

def node_c(state: state):
    print("Node C")
    return Command(
        goto=END,
        update={

            "value": state["value"]+"c"
        }
    )

def node_d(state: state): 
    print("Node D")
    return Command(
        goto=END, 
        update={
            "value": state["value"] + "d"
        }
)


graph = StateGraph(state)

graph.add_node("node_a", node_a)
graph.add_node("node_b", node_b)
graph.add_node("node_c", node_c)
graph.add_node("node_d", node_d)

graph.set_entry_point("node_a")
app = graph.compile(checkpointer=memory)


config = {"configurable": {"thread_id": "1"}}

initialState = {
    "value": ""
}

first_result = app.invoke(initialState, config, stream_mode="updates")
first_result
```

```python
print(app.get_state(config).next)
```
